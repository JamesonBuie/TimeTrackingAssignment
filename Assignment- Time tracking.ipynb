### Correlations:
- Examine whether there are any correlations between the day and the time spent. For example, do you spend more time on courses over days or less.


```python
cols=['Day','Time (min)']
heat_map=df[['Day','Time (min)']].corr()
```

```python
sns.heatmap(heat_map,xticklabels=cols,yticklabels=cols,annot=True,cmap='GnBu')
```

#### There is a fairly weak, negative linear relationship between days and time spent on work. It could possibly be said that as days go on, my time spent on a class may decrease.
